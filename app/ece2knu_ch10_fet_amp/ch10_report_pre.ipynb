### 10.4.1 그림 10.4(b) CS증폭기회로1회로에 대하여, R_2 변화(R_2 = 1.8 MegOhm, 2 MegOhm, 2.2 MegOhm)에 따른 동작점을 NgSpice로 직류해석을 하시오.

* V_DD = 10 V
* R_1 = 10 MOhm
* R_D = 2 kOhm

* 사용 소자
    + 2N7000: Place Symbols > Simulation_SPICE > NMOS
        - Simulation Model: M2N7000.lib > M2N7000

R2 가변 시험 코드

.param Rvar=1.8Meg

.control
    echo "x1 i(xq1:d) x2 v(/v_g) x3 v(/v_d)"  > ../../data/ch10_04_01_result_sim.txt

    foreach R 1.8Meg 2Meg 2.2Meg
        alterparam Rvar = $R
        reset
        save i(xq1:d) v(/v_g) v(/v_d)

        op

        echo "Step Information:" >> ../../data/ch10_04_01_result_sim.txt
        set appendwrite
        wrdata ../../data/ch10_04_01_result_sim.txt i(xq1:d) v(/v_g) v(/v_d)
    end
.endc

시뮬레이션 결과 분석

```python
R2_Ohm = np.array([1_800_000, 2_00_000, 2_200_000])

# 실험 결과 파일을 읽는다.
data = get_simulation_result('data/ch10_04_01_result_sim.txt')

I_D_mA_l = []
V_G_V_l = []
V_D_V_l = []

# 실험 결과에서 시간, 입력 전압, 출력 전압을 가져온다.
for idx in range(len(R2_Ohm)):
    I_D_mA_l.append(data['(I(XQ1:D))@%d' % (idx+1)])
    V_G_V_l.append(data['(V(/V_G))@%d'   % (idx+1)])
    V_D_V_l.append(data['(V(/V_D))@%d'   % (idx+1)])

I_D_mA = np.array([x.item() for x in I_D_mA_l])
V_G_V = np.array([x.item() for x in V_G_V_l]) / 1000.
V_D_V = np.array([x.item() for x in V_D_V_l]) / 1000.
V_S_V = np.array([0., 0., 0.])
V_DS_V = V_D_V - V_S_V

print_array("R2_Ohm", R2_Ohm)
print_array("I_D_mA", I_D_mA)
print_array("V_G_V", V_G_V)
print_array("V_D_V", V_D_V)
print_array("V_DS_V", V_DS_V)

plt.xlabel('V_DS_V')
plt.ylabel('I_D_mA')
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.xlim(-0.3, 10)
plt.ylim(-0.3, 10)
for idx in range(len(R2_Ohm)):
    label = 'R2_Ohm=%d' % R2_Ohm[idx]
    draw_plot(V_DS_V[idx], I_D_mA[idx], label, idx+1)
plt.show()
```
